```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset into a NetworkX graph
G = nx.Graph()
```

```python
# Read the dataset from the provided text file
with open('/Users/amenhasfaw/Downloads/com-amazon.top5000.cmty.txt', 'r') as file:
    for line in file:
        # Split the line into individual node IDs
        nodes = line.strip().split('\t')
        # Add edges between nodes
        for i in range(len(nodes)):
            for j in range(i + 1, len(nodes)):
                G.add_edge(nodes[i], nodes[j])
```

```python
# Calculate basic network statistics
print("Number of nodes:", nx.number_of_nodes(G))
print("Number of edges:", nx.number_of_edges(G))
```

```python
# Calculate network metrics
degree_centrality = nx.degree_centrality(G)
clustering_coefficient = nx.average_clustering(G)
betweenness_centrality = nx.betweenness_centrality(G)
pagerank = nx.pagerank(G)

print("Average Clustering Coefficient:", clustering_coefficient)
```

```python
# Visualization: Draw the network graph with edges and weights
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, node_size=50) 
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Network Graph")
plt.show()
```
